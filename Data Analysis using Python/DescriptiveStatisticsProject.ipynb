Q6)  Suppose you have been told to measure the average height of all the males on the earth. What
would be your strategy for the same? Would the average height be a parameter or a statistic?
Justify your answer

A6) To measure the average height of all the males on earth, we have to device a comprehensive strategy because of the vastness and diversity of the population. 
It is practically impossible to measure average height by including every male on earth, so we should take a representative sample which includes diferent region, country, ethnicity to avoid bias.
after taking the sample, we measure the height of each male perfectly by using tools like height measurement scales or any othe r device that gives exact height.
After measuring height of every sample from the sample population, now find the average height of the sample as follows::

Average height of all males on earth =  sum of all the heights / no of males

Parameter or Statistic ???

Parameter: If ywe aim to generalize the average height obtained from the sample to the entire population of males on Earth, it would be considered a parameter. However, obtaining the true population parameter would require measuring the height of every male on Earth, which is not feasible.
Statistic: The average height calculated from the sample is a statistic because it represents a numerical summary of a subset of the population (the sampled males), not the entire population.
Justification for the question::
The average height calculated from the sample would be a STATISTIC rather than a parameter because it is based on data collected from a subset (sample) of the population (males on Earth), not the entire population itself. Parameters are characteristics of the entire population, which are typically unknown or impractical to measure directly, while statistics are calculated from sample data and used to estimate population parameters.

```python
# 17) 17. Give the IQR of all the variables which are quantitative and continuous.

# Identifying quantitative and continuous variables
quantitative_continuous = [col for col in BankPersonal_Loan.columns if 
                             pd.api.types.is_numeric_dtype(BankPersonal_Loan[col]) and not pd.api.types.is_integer_dtype(BankPersonal_Loan[col])]

# Function to calculate IQR
def calculate_iqr(data_series):
  """
  This function calculates the IQR (Interquartile Range) for a given data series.

  Args:
      data_series (pandas.Series): The data series for which to calculate IQR.

  Returns:
      float: The IQR value.
  """
  quartiles = data_series.quantile([0.25, 0.75])
  return quartiles[0.75] - quartiles[0.25]

# Calculate and print IQR for each quantitative and continuous variable
for var in quantitative_continuous:
  iqr = calculate_iqr(BankPersonal_Loan[var])
  print(f"Variable: {var}, IQR: {iqr}")
```
